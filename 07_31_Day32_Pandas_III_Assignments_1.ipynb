```python
t = """CREATE TABLE `FitBit_data` (

`Id` DECIMAL NOT NULL, 
`ActivityDate` DATE NOT NULL, 
`TotalSteps` DECIMAL NOT NULL, 
`TotalDistance` DECIMAL NOT NULL, 
`TrackerDistance` DECIMAL NOT NULL, 
`LoggedActivitiesDistance` DECIMAL NOT NULL, 
`VeryActiveDistance` DECIMAL NOT NULL, 
`ModeratelyActiveDistance` DECIMAL NOT NULL, 
`LightActiveDistance` DECIMAL NOT NULL, 
`SedentaryActiveDistance` DECIMAL NOT NULL, 
`VeryActiveMinutes` DECIMAL NOT NULL, 
`FairlyActiveMinutes` DECIMAL NOT NULL, 
`LightlyActiveMinutes` DECIMAL NOT NULL, 
`SedentaryMinutes` DECIMAL NOT NULL, 
`Calories` DECIMAL NOT NULL
);"""
```

```python
try: 
    cursor.execute(t)
    print("Table created successfully")
except mysql.connector.Error as error:
    print("Failed to create table in MySQL: {}".format(error))
finally:
    if mydb.is_connected():
        cursor.close()
        mydb.close()
        print("MySQL connection is closed")
```
